```python
def syst_3(num):
    base = 3
    
    newNum = ''

    while num > 0:
        newNum = str(num % base) + newNum
        num //= base

    return newNum

def size(ar:np.array):
    
    ones_zeros = 0
    
    for i in range(ar.shape[0]):
        for j in range(ar.shape[1]):
            if ar[i,j] == 0 or ar[i,j] == 1:
                ones_zeros+=1
    return ones_zeros
```

```python
N = 5
list_matr = []
for i in range(3**(N)):
    mat = np.array([-1 for i in range(N*N)]).reshape((N,N))
    num = syst_3(i)
    num = '0' * (N - len(num)) + num
    
    exit = 0
    for j in range(N-1):
        
        if ((num[j] == '0' and num[j+1] == '1') or (num[j] == '1' and num[j+1] == '0')):
            exit = 1
            break
        
    if exit:
        continue
    
    
    mat[:,0] = [int(k) for k in num]
    list_matr.append(mat)
```

```python
def repaint(stack, mat, colors, max_color):
    '''
        Заходим сюда, если нашли вершину, которую не можем покрасить ни 
        в один цвет
    '''
    while len(stack) != 0:
        head = stack[-1]
        for i1 in range(head[1]+1, max_color):
            exit = 0
            for neigh in mat[head[0]]:
                if colors[neigh] == i1:
                    exit = 1

            if exit == 1:
                continue
            else:
                stack[-1][1] = i1
                colors[head[0]] = i1
                return
        colors[head[0]] = -1
        stack.pop()
```
